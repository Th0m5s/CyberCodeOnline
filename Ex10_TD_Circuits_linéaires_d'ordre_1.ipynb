**Q5.** Importer les modules numpy et matplotlib.pyplot avec les alias respectifs **np** et **plt**.





```python
# Votre code ici
import numpy as np
import matplotlib.pyplot as plt
```

**Q6.** Créer une liste T de dates comprises entre *t* = 0 s et *t* = 120 s avec un pas *p* = 0,1 s.


```python
# Votre code ici
p = 0.1
T = np.arange(1, 120.1, p)
print(T)
```

**Q7.** Créer une liste U contenant la valeur initiale de la tension aux bornes du condensateur.


```python
# Votre code ici
U=[0]
```

**Q8.** A l'aide d'une boucle **for**, remplir la liste U en ajoutant à chaque tour de boucle une nouvelle valeur de $ u_C(t) $ .


```python
# Votre code ici
for i in range(1,len(T)):
  U.append(U[i-1]+(p/20)*(30-U[i-1]))
print(U)
```

**Q9.** Ecrire le code permettant d'obtenir la figure suivante (les points sont représentés par des croix vertes) 

RC_ex10.png

```python
# Votre code ici
plt.plot(T, U, '+g')
plt.title("Charge d'un condensateur")
plt.xlabel("temps (s)")
plt.ylabel("tension aux bornes du condensateur (V)")
plt.grid()
plt.show()
```
